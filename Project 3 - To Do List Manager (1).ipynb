# âœ… To-Do List Manager
A simple task manager to:
- Add task
- View task
- Delete task
- Mark task as completed

```python
tasks = []
```

```python
def add_task():
    task = input("Enter new task: ")
    tasks.append({"task":task,"status":"Pending"})
```

```python
def view_tasks():
    for i,t in enumerate(tasks):
        print(i+1, t["task"], "-", t["status"])
```

```python
def complete_task():
    view_tasks()
    num = int(input("Task number to complete: "))
    tasks[num-1]["status"] = "Completed"
```

```python
def delete_task():
    view_tasks()
    num = int(input("Task number to delete: "))
    tasks.pop(num-1)
```

```python
def run_todo():
    while True:
        print("\n1.Add 2.View 3.Complete 4.Delete 5.Exit")
        ch = input("Choice: ")
        
        if ch=='1': add_task()
        elif ch=='2': view_tasks()
        elif ch=='3': complete_task()
        elif ch=='4': delete_task()
        elif ch=='5': break
```

```python
run_todo()
```
